```python
import pandas as pd
import warnings
```

```python
# читаем данные из таблицы аспо деталиазция.xlsx
wb = pd.read_excel('аспо деталиазция.xlsx', sheet_name='Sheet1')
wb['Артикул поставщика'] = wb['Артикул поставщика'].astype('string')
```

Для суммирования данных нам необходимо поменять положительные на отрицательные и наоборот значения в таблице по некоторым колонкам при условии 'Тип документа' == 'Возврат':

```python
# К перечислению за товар
wb.loc[wb['Тип документа'] == 'Возврат', 'Вайлдберриз реализовал Товар (Пр)'] = -wb['Вайлдберриз реализовал Товар (Пр)']
```

```python
# комисся вайлдберис, НДС комиссии
wb.loc[wb['Тип документа'] == 'Возврат', 'Вознаграждение Вайлдберриз (ВВ), без НДС'] = -wb['Вознаграждение Вайлдберриз (ВВ), без НДС']
wb.loc[wb['Тип документа'] == 'Возврат', 'НДС с Вознаграждения Вайлдберриз'] = -wb['НДС с Вознаграждения Вайлдберриз']
```

```python
# Возмещение расходов поверенного 
wb.loc[wb['Тип документа'] == 'Возврат', 'Возмещение за выдачу и возврат товаров на ПВЗ'] = -wb['Возмещение за выдачу и возврат товаров на ПВЗ']
wb.loc[wb['Тип документа'] == 'Возврат', 'Возмещение издержек по эквайрингу'] = -wb['Возмещение издержек по эквайрингу']
```

<hr>

### Добавление данных о хранении  
Оплату за хранение мы знаем только итоговой суммой.  
Мы будем делить сумму стоимости хранения на долю стоимости каждого товара от выручки.

```python
# сумма стоимости размещения всех товаров за указанный период
warehousing = 17943.48

# сумма всей выручки
revenue = wb[(wb['Тип документа'] == 'Продажа') & (wb['Обоснование для оплаты'] == 'Продажа')] \
    ['Вайлдберриз реализовал Товар (Пр)'].sum()

# доля хранения от выручки
storage = warehousing / revenue

# добавляем колонку
wb['Стоимость хранения'] = 0

# добавляем в таблицу данные о хранении только в строки о продаже
wb.loc[(wb['Тип документа'] == 'Продажа') & (wb['Обоснование для оплаты'] == 'Продажа'), 'Стоимость хранения'] = wb['Вайлдберриз реализовал Товар (Пр)'] * storage
```

<hr>

### Добавление данных о рекламе
Название рекламных компаний нам ничего не говорят о конкретном товаре, который мы рекламируем.  
Для расчета мы можем только приблизительно понять к какой категории относится рекламная компания.  
После этого мы суммируем рекламные расходы по категориям, а затем разбиваем эти суммы на каждый товар в зависимости от категории, аналогичным способом Стоимости хранения.

```python
# создаём столбец с данными о рекламе
wb['Реклама'] = 0
```

```python
# получаем данные из таблиц о рекламных компаниях
ad_1 = pd.read_excel('Расходы на рекламу.xlsx', sheet_name='Карточка товара')
ad_2 = pd.read_excel('Расходы на рекламу.xlsx', sheet_name='Каталог')
ad_3 = pd.read_excel('Расходы на рекламу.xlsx', sheet_name='Поиск')

ad_1
```

```python
# объединяем данные с разных типов рекламы и группируем по категориям товаров
advertising = pd.concat([ad_1[['Предмет', 'Сумма, р']], ad_2[['Предмет', 'Сумма, р']], ad_3[['Предмет', 'Сумма, р']]])
advertising = advertising.groupby('Предмет').agg({'Сумма, р': sum}).reset_index()
advertising
```

```python
# добавим рекламные расходы к каждому товару в общую таблицу
for item in advertising['Предмет'].tolist():
    revenue = wb[(wb['Предмет'] == item) & (wb['Тип документа'] == 'Продажа') & (wb['Обоснование для оплаты'] == 'Продажа')] \
        ['Вайлдберриз реализовал Товар (Пр)'].sum()
    warehousing = advertising[advertising['Предмет'] == item]['Сумма, р'].values[0]
    storage = warehousing / revenue
    wb.loc[(wb['Предмет'] == item) & (wb['Тип документа'] == 'Продажа') & (wb['Обоснование для оплаты'] == 'Продажа'), 'Реклама'] = wb['Вайлдберриз реализовал Товар (Пр)'] * storage
```

### Добавим данных о себестимости товара, и стоимости логистики до ваилдберис

Мы могли бы обратится к таблице "Фин модель" но оказалось что она не содержит точных данных об артикулах товаров с которыми можно было бы сопоставить себестоимость товара и стоимость доставки до ваилдберис.

Пришлось наугад примерно выбрать подходящие названиям модели, дописать к нам артикулы товаров и пересобрать данные в таблицу 'Фин_модель_ред.xlsx'

После этого мы можем добавить эту информацию к общим данным.

```python
# получим данные о рекламе
finmod = pd.read_excel('Фин_модель_ред.xlsx')[['Артикул поставщика', 'Себ-ть товара на нашем складе', 'Доставка до WB']]
finmod['Артикул поставщика'] = finmod['Артикул поставщика'].astype('string')
finmod
```

```python
# удалим лишнее, в сучае если это не продажи и возвраты
wb.loc[(wb['Тип документа'] == 'Продажа') & (wb['Обоснование для оплаты'] != 'Продажа') | \
    (wb['Тип документа'] == 'Возврат') & (~wb['Обоснование для оплаты'].isin(['Возврат', 'Авансовая оплата за товар без движения'])), \
       'Себ-ть товара на нашем складе'] = 0

wb.loc[(wb['Тип документа'] == 'Продажа') & (wb['Обоснование для оплаты'] != 'Продажа') | \
    (wb['Тип документа'] == 'Возврат') & (~wb['Обоснование для оплаты'].isin(['Возврат', 'Авансовая оплата за товар без движения'])), \
       'Доставка до WB'] = 0
```

### Сводная таблица

После того как мы собрали и объединили все данные мы можем:
- отобрать только интересующие нас позиции
- сгруппировать данные по каждому товару за веь период
- посчитать что нам перечислит ваилдберис в колонку 'Итого к перечислению продавцу'
- посчитать чистую прибыль

```python
# оставим только необходимые для расчёта колонки
wb = wb[['Код номенклатуры', 
         'Артикул поставщика',
         'Предмет',
         'Название', 
         'Тип документа', 
         'Кол-во',
         'Обоснование для оплаты', 
         'Вайлдберриз реализовал Товар (Пр)', 
         'Вознаграждение Вайлдберриз (ВВ), без НДС', 
         'НДС с Вознаграждения Вайлдберриз', 
         'Услуги по доставке товара покупателю',
         'Возмещение за выдачу и возврат товаров на ПВЗ', 
         'Возмещение издержек по эквайрингу', 
         'Общая сумма штрафов',
         'Виды логистики, штрафов и доплат', 
         'Стоимость хранения',
         'Реклама',
         'ШК', 
         'Себ-ть товара на нашем складе', 
         'Доставка до WB']]
```

```python
# определяем товары, которые нас интересуют
goods_types = ['Мотоперчатки', 'Мотоштаны', 'Мотокуртка', 'Мотошлем', 'Мототент']

# оставляем только выбрнные товары
wb['categ'] = wb['Название'].apply(lambda x: x.split(' ')[0])
wb = wb[wb.categ.isin(goods_types)]

# также удаляем тактические перчатки
wb = wb[wb['Код номенклатуры'] != 35302428]

# Будем работать по этим товарам
wb.sort_values('Название')['Название'].unique()
```

```python
# сгруппируем данные по выбранным товарам
group = wb.groupby('Название').agg({'Артикул поставщика': 'first', 
                                    'Кол-во': sum, 
                                    'Вайлдберриз реализовал Товар (Пр)': sum, 
                                    'Вознаграждение Вайлдберриз (ВВ), без НДС': sum, 
                                    'НДС с Вознаграждения Вайлдберриз': sum, 
                                    'Услуги по доставке товара покупателю': sum,
                                    'Возмещение за выдачу и возврат товаров на ПВЗ': sum, 
                                    'Возмещение издержек по эквайрингу': sum, 
                                    'Общая сумма штрафов': sum, 
                                    'Стоимость хранения': sum, 
                                    'Реклама': sum,
                                    'Себ-ть товара на нашем складе': sum, 
                                    'Доставка до WB': sum})
group.reset_index(inplace= True)
```

```python
# считаем Итого к перчислению продавцу
group['Итого к перечислению продавцу'] = group['Вайлдберриз реализовал Товар (Пр)'] \
    - group['Вознаграждение Вайлдберриз (ВВ), без НДС'] \
    - group['НДС с Вознаграждения Вайлдберриз'] \
    - group['Услуги по доставке товара покупателю'] \
    - group['Стоимость хранения'] \
    - group['Возмещение за выдачу и возврат товаров на ПВЗ'] \
    - group['Возмещение издержек по эквайрингу'] \
    - group['Общая сумма штрафов'] \
    - group['Реклама']

# считаем все затраты
group['Все затраты'] = group['Вознаграждение Вайлдберриз (ВВ), без НДС'] \
    + group['НДС с Вознаграждения Вайлдберриз'] \
    + group['Услуги по доставке товара покупателю'] \
    + group['Стоимость хранения'] \
    + group['Возмещение за выдачу и возврат товаров на ПВЗ'] \
    + group['Возмещение издержек по эквайрингу'] \
    + group['Общая сумма штрафов'] \
    + group['Реклама'] \
    + group['Себ-ть товара на нашем складе'] \
    + group['Доставка до WB']

# прибыль
group['Прибыль'] = (group['Вайлдберриз реализовал Товар (Пр)'] - group['Все затраты'])

# считаем ROI
group['ROI'] = round(group['Прибыль'] / group['Все затраты'], 2)
```

```python
# переставляем колонки
group = group[['Артикул поставщика', 
        'Название',
       'Кол-во', 
        'Вайлдберриз реализовал Товар (Пр)', 
       'Вознаграждение Вайлдберриз (ВВ), без НДС', 
       'НДС с Вознаграждения Вайлдберриз', 
       'Услуги по доставке товара покупателю', 
       'Стоимость хранения', 
       'Возмещение за выдачу и возврат товаров на ПВЗ', 
       'Возмещение издержек по эквайрингу', 
       'Общая сумма штрафов', 
       'Реклама', 
        'Итого к перечислению продавцу', 
               
       'Себ-ть товара на нашем складе', 
       'Доставка до WB', 
    
       'Все затраты',
       'Прибыль',
       'ROI']]

group['Стоимость хранения'] = round(group['Стоимость хранения'], 2)
group['Реклама'] = round(group['Реклама'], 2)
group['Итого к перечислению продавцу'] = round(group['Итого к перечислению продавцу'], 2)
group['Себ-ть товара на нашем складе'] = round(group['Себ-ть товара на нашем складе'], 2)
group['Доставка до WB'] = round(group['Доставка до WB'], 2)
group['Вайлдберриз реализовал Товар (Пр)'] = round(group['Вайлдберриз реализовал Товар (Пр)'], 2)
group['Все затраты'] = round(group['Все затраты'], 2)
group['Прибыль'] = round(group['Прибыль'], 2)

group
```
